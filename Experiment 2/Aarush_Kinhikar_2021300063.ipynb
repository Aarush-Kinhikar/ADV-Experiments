```python
for column in df.columns.drop(['Country','Currency']):
    # Sort and get top 10 countries for the column
    top_10_countries = df_2021[['Country', column]].sort_values(by=column, ascending=False).head(10)

    # Plotting
    plt.figure(figsize=(6, 4))
    sns.barplot(x=column, y='Country', data=top_10_countries, palette='viridis')
    plt.title(f'Top 10 Countries by {column} in 2021')
    plt.xlabel(column)
    plt.ylabel('Country')
    plt.show()
```

```python
for column in df.columns.drop(['Country','Currency']):
    # Sort and get bottom 10 countries for the column
    bottom_10_countries = df_2021[['Country', column]].sort_values(by=column, ascending=False).tail(10)

    # Plotting
    plt.figure(figsize=(6, 4))
    sns.barplot(x=column, y='Country', data=bottom_10_countries, palette='viridis')
    plt.title(f'Bottom 10 Countries by {column} in 2021')
    plt.xlabel(column)
    plt.ylabel('Country')
    plt.show()
```

<h2>Line Chart For India</h2>

```python
df_India = df[df['Country'].str.strip().str.upper() == 'INDIA']

df_India.head()
```

```python
# Define a custom color palette
palette = sns.color_palette('tab10', len(df.columns.drop(['Country','Currency'])))

# Plot each column over the year
plt.figure(figsize=(20, 10))

for idx, column in enumerate(df.columns.drop(['Country','Currency'])):
    sns.lineplot(data=df_India, x='Year', y=column, marker='o', label=column, color=palette[idx])

# Customize the plot
plt.title('Metrics for India Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.legend(title='Metric', title_fontsize='13', fontsize='11', loc='upper left')  # Adjust the legend font size and position
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()  # Adjusts plot to ensure everything fits without overlapping

plt.show()
```

<h2>Stacked Area Chart For India</h2>

```python
plt.figure(figsize=(14, 8))

# Data for stacked area chart
years = df_India['Year']
data_to_stack = [df_India[column] for column in df.columns.drop(['Country','Currency'])]

plt.stackplot(years, data_to_stack, labels=df.columns.drop(['Country','Currency']), alpha=0.6)

plt.title('Metrics for India Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.legend(title='Metric', title_fontsize='13', fontsize='11', loc='upper left')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract relevant columns for the doughnut chart
sector_labels = [
    'Agriculture, hunting, forestry, fishing (ISIC A-B)',
    'Mining, Manufacturing, Utilities (ISIC C-E)',
    'Construction (ISIC F)',
    'Wholesale, retail trade, restaurants and hotels (ISIC G-H)',
    'Transport, storage and communication (ISIC I)',
    'Other Activities (ISIC J-P)'
]
sector_values = [country_data[col].values[0] for col in sector_labels]

# Plotting the doughnut chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(sector_values, labels=sector_labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors, wedgeprops={'edgecolor': 'black'})

# Add a central circle for the doughnut effect
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Title and display
ax.set_title('Doughnut Chart - GDP Breakdown by Sector')
plt.show()
```

```python
import plotly.graph_objects as go

# Filter data for a specific country and year
country_data = df[df['Country'].str.strip().str.upper() == 'INDIA']
country_data = country_data[country_data['Year'] == 2021]

# Extract relevant columns for the waterfall chart
sectors = [
    'Agriculture, hunting, forestry, fishing (ISIC A-B)',
    'Mining, Manufacturing, Utilities (ISIC C-E)',
    'Construction (ISIC F)',
    'Wholesale, retail trade, restaurants and hotels (ISIC G-H)',
    'Transport, storage and communication (ISIC I)',
    'Other Activities (ISIC J-P)'
]

sector_values = [country_data[col].values[0] for col in sectors]
gdp_value = country_data['Gross Domestic Product (GDP)'].values[0]

# Prepare data for the waterfall chart
sector_values.append(gdp_value)  # Add GDP as the final total
sectors.append('Gross Domestic Product (GDP)')

# Create the waterfall chart
fig = go.Figure(go.Waterfall(
    x=sectors,
    y=sector_values,
    measure=["relative"] * (len(sectors) - 1) + ["total"],
    connector=dict(line=dict(color="rgb(63, 63, 63)", width=2))
))

fig.update_layout(
    title="Waterfall Chart - Sectoral Contributions to GDP",
    xaxis_title="Economic Sector",
    yaxis_title="Contribution to GDP (in million USD)"
)

fig.show()
```
